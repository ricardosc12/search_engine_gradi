```python
import pandas as pd
from ast import literal_eval

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

db_games = pd.read_csv("./games.csv")

db_games.head(4)
```

```python
db_games.drop('Average playtime forever', axis=1, inplace=True)
db_games.drop('Average playtime two weeks', axis=1, inplace=True)
db_games.drop('Median playtime forever', axis=1, inplace=True)
db_games.drop('Median playtime two weeks', axis=1, inplace=True)
db_games.drop('Website', axis=1, inplace=True)
db_games.drop('Support url', axis=1, inplace=True)

db_games.drop('Movies', axis=1, inplace=True)
db_games.drop('Tags', axis=1, inplace=True)
db_games.drop('Achievements', axis=1, inplace=True)
db_games.drop('Metacritic url', axis=1, inplace=True)
db_games.drop('Support email', axis=1, inplace=True)
db_games.drop('Reviews', axis=1, inplace=True)
db_games.drop('Peak CCU', axis=1, inplace=True)
db_games.drop('DLC count', axis=1, inplace=True)
db_games.drop('Full audio languages', axis=1, inplace=True)
db_games.drop('Score rank', axis=1, inplace=True)
db_games.drop('Notes', axis=1, inplace=True)
```

```python
db_games.rename(columns={'Name':'name'}, inplace=True)
db_games.rename(columns={'Release date':'release_date'}, inplace=True)
db_games.rename(columns={'Estimated owners':'estimated_owners'}, inplace=True)
db_games.rename(columns={'Required age':'required_age'}, inplace=True)
db_games.rename(columns={'Price':'price'}, inplace=True)
db_games.rename(columns={'About the game':'description'}, inplace=True)
db_games.rename(columns={'Supported languages':'languages'}, inplace=True)
db_games.rename(columns={'Header image':'image'}, inplace=True)
db_games.rename(columns={'Windows':'windows'}, inplace=True)
db_games.rename(columns={'Mac':'mac'}, inplace=True)
db_games.rename(columns={'Linux':'linux'}, inplace=True)
db_games.rename(columns={'Metacritic score':'meta_score'}, inplace=True)
db_games.rename(columns={'User score':'user_score'}, inplace=True)
db_games.rename(columns={'Positive':'positive'}, inplace=True)
db_games.rename(columns={'Negative':'negative'}, inplace=True)
db_games.rename(columns={'Recommendations':'recommendations'}, inplace=True)
db_games.rename(columns={'Developers':'developers'}, inplace=True)
db_games.rename(columns={'Publishers':'publishers'}, inplace=True)
db_games.rename(columns={'Categories':'categories'}, inplace=True)
db_games.rename(columns={'Genres':'genres'}, inplace=True)
db_games.rename(columns={'Screenshots':'screenshots'}, inplace=True)
```

```python
db_games['release_date'] = pd.to_datetime(db_games['release_date'])
# db_games['languages'] = db_games['languages'].apply(lambda x: ', '.join(literal_eval(x)))

def convert_to_list(lang_str):
    languages = lang_str.strip("[]").split("', '")
    languages[0] = languages[0].strip("'")
    languages[-1] = languages[-1].strip("'")
    return ', '.join(languages)

db_games['languages'] = db_games['languages'].apply(convert_to_list)
```

```python
db_games.to_csv('game_database.csv' ,index=False)
```
